```python
from sklearn.model_selection import train_test_split
pr_train1, pr_test1 = train_test_split(pr_train, test_size = 0.2,random_state=2)
```

```python
x_train1=pr_train1.drop(['Junk'],1)
y_train1=pr_train1['Junk']
x_test1=pr_test1.drop(['Junk'],1)
y_test1=pr_test1['Junk']
```

```python
x_train1.shape,y_train1.shape
```

```python
base_clf=DecisionTreeClassifier(max_leaf_nodes=10,class_weight='balanced')
```

```python
adb_params={'n_estimators':[50,100,200,500,700],
           'learning_rate': [0.01,.1,1]
           }
adb=AdaBoostClassifier(base_estimator=base_clf)
```

```python
complete_search=GridSearchCV(adb,scoring='roc_auc',param_grid=adb_params,cv=10,n_jobs=-1,verbose=2)
```

```python
complete_search.fit(x_train1,y_train1)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.5f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
report(complete_search.cv_results_,5)
```
